```python
phones_to_category = []

for payer in range(len(array)):
    d = dict()
    count = 0
    for transfer in range(len(array[payer])):
        if array[payer][transfer][2] not in d.keys():
            d[array[payer][transfer][2]] = count
            count += 1
    phones_to_category.append(d)
print(phones_to_category)
```

```python
final_array_input = []
final_array_train = []
final_array_test = []
final_phones_to_category = []
for payer in range(len(array)):
    d = dict()
    count = 0
    try:
        for transfer in range(len(array[payer])):
            array[payer][transfer][2] = phones_to_category[payer][array[payer][transfer][2]]

        array_train[payer] = phones_to_category[payer][array_train[payer]]
        array_test[payer] = phones_to_category[payer][array_test[payer]]
        final_array_input.append(array[payer])
        final_array_train.append(array_train[payer])
        final_array_test.append(array_test[payer])
        final_phones_to_category.append(phones_to_category[payer])
    except KeyError as e:
        # print(phones_to_category[payer])
        # print(array_train[payer])
        # print(array_test[payer])
        # print('-------------')
        # print(e)
        pass

print(len(final_array_input))
```

```python
from collections import Counter

success_count = 0
fail_count = 0

only_input_numbers = []
# print(final_array_train)
for i in range(len(final_array_input)):
    only_input_numbers.append([final_array_input[i][j][2] for j in range(len(final_array_input[i]))])
    only_input_numbers[i] = list(dict.fromkeys(only_input_numbers[i])) # remove duplicates
# print(only_input_numbers[0])
for payer in range(len(final_array_input)):
# for payer in range(2):
#     print(final_array_train[payer], random.sample(only_input_numbers[payer], 3))
    try:
        # print(Counter(only_input_numbers[payer]).most_common(3))

        frequency_sample = [Counter(only_input_numbers[payer]).most_common(3)[i][0] for i in range(3)]
        if final_array_train[payer] in frequency_sample:
            # print(len(only_input_numbers[payer]), final_array_train[payer], random_sample)
            success_count += 1
        else:
            fail_count += 1
    except IndexError as e:
        success_count += 1

print("FREQUENCY TOP3 SUCCESS RATE =", success_count/(success_count+fail_count))
```
